```python
#Filter and save coordinates in out.txt file

gff_file = 'watermelon.gff'
fasta_file = 'watermelon.fsa'
out_file = "out.txt"

gff = open(gff_file, 'r')
fasta = open(fasta_file, 'r')
output = open(out_file, "w")
for line in gff:

    fields = line.split('\t')
    start = int(fields[3])
    stop = int(fields[4])
    print(str(start) + "\t" + str(stop) + '\n')
    output.write(str(start) + "\t" + str(stop) + '\n')
```

```python
# Find substrings

fasta_file = open("watermelon.fsa").read()
out_file = open("out.txt")

output = open("substring_sequence.txt", "w")

for line in out_file:
    positions = line.split('\t')
    start = int(positions[0])
    stop = int(positions[1])
    substring = fasta_file[start:stop]
    print(substring)
    output.write(substring)
    
    
```

```python
# Find GC content in the substring file

from __future__ import division
substring_file = open("substring_sequence.txt").read()
Gnum = substring_file.count('G')
Cnum = substring_file.count('C')
print(100*(Gnum + Cnum)/len(substring_file))
```
